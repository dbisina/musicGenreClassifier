**IMPORTING LIBRARIES**

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
#import sklearn.external.joblib as extjoblib
import joblib
from sklearn import tree
music_data = pd.read_csv('music.csv')
```

**INPUT SET**

```python
X = music_data.drop(columns=['genre'])
```

**OUTPUT SET**

```python
y = music_data['genre']
```

# **Building a model**

```python
#calculating the accuracy
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
model = DecisionTreeClassifier()
model.fit(X_train, y_train)

#saving a model
joblib.dump(model, 'machine_learning.joblib')

#loading a saved model
#model = joblib.dump(model, 'machine_learning.joblib')

predictions = model.predict(X_test)

score = accuracy_score(y_test, predictions)
```

** Visualizing a Decision tree**

```python
tree.export_graphviz(model, out_file = 'machine_leasrning.dot', feature_names=['age','gender'], class_names=sorted(y.unique()), label ='all',rounded=True, filled=True )
```
